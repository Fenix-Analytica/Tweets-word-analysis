# Conteo de palabras y Wordcloud

```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
import nltk
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_excel("DBTweets.xlsx")
df.head(5)
```

```python
text=df.AT
x=text[0]
for j in range (1,len(text)):
    temp=text[j]
    esp=" "
    x=x+esp+temp
x=x.lower()
```

```python
from nltk.tokenize import RegexpTokenizer
x=re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', x)
wordcloud = WordCloud().generate(x)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
tokenizer = RegexpTokenizer(r'\w+')
q=tokenizer.tokenize(x)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(q)
fdist.most_common(1)
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("spanish"))
filtered_sent=[]
for w in q:
    if w not in stop_words:
        filtered_sent.append(w)
q2=' '.join([str(elem) for elem in filtered_sent])
wordcloud = WordCloud().generate(q2)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
